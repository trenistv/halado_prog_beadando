```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn import linear_model

print("1 - Fájlból való beolvasás")
print("2 - Generált számok használata")
print("3 - Saját magam adom meg")

def int_lin_reg(ans):
  match ans:
    case '1':
      fajlnev = input("Adja meg a fájl nevét: ")
      adat = np.loadtxt(fajlnev, delimiter=';', skiprows=1)
      x = adat[:,0].reshape(-1,1)
      y = adat[:,1]

    case '2':
      print("Random adatok generálása...")
      x = np.random.uniform(1, 10, 30).reshape(-1,1)

      zaj = np.random.normal(0, 1, 30).reshape(-1,1)
      y = 2.5 * x + zaj
      y = y.flatten()

    case '3':
        x_lista = []
        y_lista = []

        print("Adja meg az X,Y értékeket! (Pl: 3, 7)")
        print("Írjon 'stop'-ot a befejezéshez.")

        while True:
            sor = input("X,Y = ")
            if sor.lower() == "stop":
                break
            try:
                x_val, y_val = sor.split(",")
                x_lista.append(float(x_val))
                y_lista.append(float(y_val))
            except:
                print("Hibás formátum! Példa: 3, 7")

        x = np.array(x_lista).reshape(-1, 1)
        y = np.array(y_lista)
    case _:
      print("Hibás választás!")
      return

  model = linear_model.LinearRegression()
  model.fit(x,y)

  plt.scatter(x, y, label='Adatok')
  plt.plot(x, model.predict(x), label='Lineáris regresszió', linewidth=2)
  plt.title("Lineáris regresszió")
  plt.xlabel("X")
  plt.ylabel("Y")
  plt.legend()
  plt.show()

int_lin_reg(input("Válassza ki melyiket szeretné használni: "))
```
